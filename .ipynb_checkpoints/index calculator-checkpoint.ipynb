# Find best correction lens

```python
bas = 924
matas = db.get_material(bas)
database = np.arange(805,1067,dtype=float)
waveshort = 780
wavelong = 852
wavemid = (waveshort + wavelong)/2
f1 = 6
c1 = 1/f1/matas.get_refractiveindex(wavemid)
f3 = np.arange(-1000,-7,dtype=float)
f22=0
f33=0
vmax = 0
vmin = 10000
vmann=0
vminn=0
n = 0
n_min = 1
r = -100
for i in database:
#    print(i)
    mat = db.get_material(i)
    r = -1
    try:
        inx = mat.get_extinctioncoefficient(780)
    except Exception:
        continue
    if (inx) > 1e-7:
        continue
    try:
        mat.get_refractiveindex(400)
    except Exception:
        continue 
    try:
        mat.get_refractiveindex(1500)
    except Exception:
        continue 
    if mat.get_refractiveindex(wavemid) > 1.6:       
        for j in f3:
            f2 = 1/(1/20-1/j)
            c2 = 1/f2/(matas.get_refractiveindex(wavemid))
            c3 = 1/(j)/mat.get_refractiveindex(wavemid)
            deln1 = (matas.get_refractiveindex(waveshort)-matas.get_refractiveindex(wavelong))
            deln2 = (matas.get_refractiveindex(waveshort)-matas.get_refractiveindex(wavelong))
            deln3 = (mat.get_refractiveindex(waveshort)-mat.get_refractiveindex(wavelong))
            totab = -(c1*deln1 + c2*deln2 + c3*deln3)
            if  totab > r :
                r = totab
                f22 = c2
                f33 = c3
    if r > 0.000:
        print(mat.get_page_info(),r,c1,f22,f33,mat.get_refractiveindex(wavemid))
```
